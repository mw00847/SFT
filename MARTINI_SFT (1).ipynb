**Surface attraction of short chain polymers to the surface**

**Abstract**

Polyethylene has a lower SFE than PS or PEO. Common conception is that the lower SFE component is found at the surface, however .... this shows that the short chain polymers can be found at the surface. In blends of the same polymer (PEO), the smaller chain is found at the surface for varying concentrations. The MARTINI forcefield has been previously shown to understimate the surface tension of water.


Low molecular weight polymers have shown to favour the surface in a blend with higher weight polymers.  

P3HT has a low surface energy, can PS or PEO be allowed to form at the interface when the chains are short enough?
